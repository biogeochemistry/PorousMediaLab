```python
import matplotlib.pyplot as plt
import sys
sys.path.append('../../src')
import numpy as np
from pHcalc import *
import numpy as np
%matplotlib inline

na_moles = np.linspace(1e-8, 5.e-3, 50000)
sol_volume = 1. # Liter
phos = Acid(pKa=[2.148, 7.198, 12.375], charge=0, conc=1.e-3)
phs = []
pH = 2
for mol in na_moles:
    na = Neutral(charge=1, conc=mol/sol_volume)
    system = System(phos, na)
    system.pHsolve(guess_est=True)
    pH = system.pH 
    phs.append(pH)
plt.plot(na_moles, phs)
plt.grid(lw=0.2)
plt.ylim(0,14)
plt.xlim(0,0.005)
plt.ylabel('pH')
plt.xlabel('Na')
plt.show()
plt.show()
```

```python
phos = Acid(pKa=[2.148, 7.198, 12.319], charge=0, conc=0.01)
phs = np.linspace(0, 14, 1000)
fracs = phos.alpha(phs)
plt.plot(phs, fracs, lw=3)
plt.grid(lw=0.2)
plt.ylim(0,1)
plt.xlim(0,14)
plt.ylabel('Alpha')
plt.xlabel('pH')
plt.show()
```
